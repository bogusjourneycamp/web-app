```python
from stl import mesh
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from ipywidgets import widgets
from ipywidgets import interact, interact_manual, interactive

import numpy as np
```

```python
the_man = mesh.Mesh.from_file('/Users/gnb/BM_Logo.stl')
the_man.rotate([1, 0,0], 3*np.pi/2)
```

```python
figure = plt.figure()
axes = mplot3d.Axes3D(figure)
# Load the STL files and add the vectors to the plot

axes.add_collection3d(mplot3d.art3d.Poly3DCollection(the_man.vectors))
# Auto scale to the mesh size
scale = the_man.points.flatten(-1)
axes.auto_scale_xyz(scale, scale, scale)
# Show the plot to the screen
plt.show()
```

```python
streets = ["Man", "Esplanade", "A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L",]

scales = np.linspace(0.6, 4, len(streets))



for i, street in enumerate(streets):
    
    print(street)

    for angle in np.arange(0, 360, 7.5):

        
        hour = angle//30 
        hour += 12*(hour==0)
        minutes = ((angle%30)//7.5)*15
        time = "{}".format(int(hour))+ "__" +"{:02}".format(int(minutes))

        view_angle = angle-90


        fig = plt.figure(figsize=(5,5), frameon=False)
        ax = fig.add_subplot(111, projection='3d')
        ax.add_collection3d(mplot3d.art3d.Poly3DCollection(the_man.vectors, color='black'))
        ax.auto_scale_xyz(scale*scales[i], scale*scales[i], scale*scales[i])
        ax.view_init(0, view_angle)
        ax.set_axis_off()   
        #fig.savefig("the_man/{}_{}.svg".format(street, time), transparent=True)
        fig.savefig("the_man/{}_{}.png".format(street, time), transparent=True, bbox_inches = 'tight', pad_inches = 0)
        plt.close(fig)
    
    #print(angle,, angle-90)
```
